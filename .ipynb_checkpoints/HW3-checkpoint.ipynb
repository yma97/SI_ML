```python
import pandas as pd
import numpy as np
```

```python
# Load the data
df = pd.read_csv("data.csv", header = None)
df
```

```python
# Get the last column as the true clusters 
true_cluster = df[df.columns[-1]]
T = np.array(df.iloc[:,:-1])
```

```python
n_d = T.shape[0]
n_w = T.shape[1]
n_c = 4
```

```python
# Initialize parameters
def initialize_random_params():
    gamma = np.ones((n_d, n_c)) / n_c 
    mu = np.random.normal(0, 1, size=(n_w, n_c))
    pi = gamma.sum(axis=0) / n_d
    
    return gamma, mu, pi
```

```python
gamma, mu, pi = initialize_random_params()
```

```python
def update_gamma(T, mu, pi):
    gamma = np.zeros((n_d, n_c))
    for i in range(n_d):
        for c in range(n_c):
            nom = 1
            denom = 0
            for j in range(n_w):
                nom *= np.power(mu[j][c],T[i][j])
            nom *= pi[c] 
            for sc in range(n_c):
                temp=1
                for j in range(n_w):
                    temp *= np.power(mu[j][sc],T[i][j])
                denom += (pi[sc] * temp)
            gamma[i][c] = nom / denom
    return gamma

def update_mu(T, gamma): 
    nom = np.dot(T.T, gamma)
    denom = nom.sum(axis=0)
    mu = nom/denom
    return mu

def update_pi(gamma):
    pi = gamma.sum(axis=0) / n_d
    return pi
```

```python
# E step 
def e_step(T, mu, pi):
    return update_gamma(T, mu, pi)

# M step
def m_step(T, gamma):
    return update_mu(T, gamma), update_pi(gamma)

#def compute_ll(T, gamma, mu, pi):
#    ll = 
#    return np.mean(loglikelihood)
```

```python
# Run em
max_iter = 20
count = 0
pred = np.zeros(n_d)
gamma, mu, pi = initialize_random_params()
while(count < max_iter):
    print(count)
    # Run e_step and m_step, update the params
    gamma = e_step(T, mu, pi)
    #print(gamma)
    mu, pi = m_step(T, gamma)
    # Get the prediction of the cluster as the index of the max probability
    for i in range(len(gamma)): # row_i = 1 * 4
        row_i = list(gamma[i])
        pred[i] = row_i.index(max(row_i))+1
    print(pred)
    count+=1
```

```python
# Report accuracy
count = 0
for i in range(len(true_cluster)):
    if true_cluster[i] == pred[i]:
        count+=1
        
accuracy = count/400
print(accuracy)
```
