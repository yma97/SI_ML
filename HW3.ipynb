```python
import pandas as pd
import numpy as np
```

```python
# Load the data
df = pd.read_csv("data.csv", header = None)
df
```

```python
# Get the last column as the true clusters 
true_cluster = df[df.columns[-1]]
T = df.iloc[:,:-1]
```

```python
n_d = T.shape[0]
n_w = T.shape[1]
n_c = 4
```

```python
# Initialize parameters
def initialize_random_params():
    gamma = np.ones((n_d, n_c)) / n_c 
    mu = np.random.normal(0, 1, size=(n_w, n_c))
    pi = gamma.sum(axis=0) / n_d
    
    return gamma, mu, pi
```

```python
def update_gamma(T, mu, pi):
    gamma = np.zeros((n_d, n_c))
    for i in range(n_d):
        for c in range(n_c):
            nom = 1
            denom = 0
            temp=1
            for j in range(n_w):
                nom *= np.power(mu[j][c],T[j][i])
            nom *= pi[c] 
            for sc in range(n_c):
                for j in range(n_w):
                    temp *= np.power(mu[j][sc],T[j][i])
                denom += (pi[sc] * temp)
            gamma[i][c] = nom / denom
    return gamma

def update_mu(T, gamma):
    mu = np.zeros((n_w, n_c))   
    nom = np.dot(T.T, gamma)
    denom = np.zeros((1,4))
    for j in range(n_w):
        denom += nom[j]
    mu = nom/denom
    
    return mu

def update_pi(gamma):
    pi = gamma.sum(axis=0) / n_d
    return pi
```

```python
# E step 
def e_step(T, mu, pi):
    return update_gamma(T, mu, pi)

# M step
def m_step(T, gamma):
    return update_mu(T, gamma), update_pi(gamma)

#def compute_ll(T, gamma, mu, pi):
#    ll = 
#    return np.mean(loglikelihood)
```

```python
# Run em
max_iter = 30
count = 0
pred = np.zeros(n_d)
gamma, mu, pi = initialize_random_params()
while(count < max_iter):
    print(count)
    # Run e_step and m_step, update the params
    gamma = e_step(T, mu, pi)
    print(gamma)
    mu, pi = m_step(T, gamma)
    # Get the prediction of the cluster as the index of the max probability
    for i in range(len(gamma)): # row_i = 1 * 4
        row_i = list(gamma[i])
        pred[i] = row_i.index(max(row_i))+1
    count+=1
```
